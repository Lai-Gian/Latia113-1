```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the DataFrame
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

# Plotting a bar chart for column 'A'
plt.figure(figsize=(10, 6))
df['A'].plot(kind='bar', color='skyblue')
plt.title('Bar Chart for Column A')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()

# Plotting a pie chart for the last 10 values of column 'B'
plt.figure(figsize=(8, 8))
df['B'].tail(10).plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'orange', 'lightcoral', 'cyan', 'yellow', 'pink', 'grey', 'purple', 'brown', 'blue'])
plt.title('Pie Chart for Last 10 Values of Column B')
plt.ylabel('')
plt.show()

# Plotting a 3D bar chart
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
x = np.arange(len(df))
y = df.columns
x_pos, y_pos = np.meshgrid(x, np.arange(len(y)))
x_pos = x_pos.flatten()
y_pos = y_pos.flatten()
z_pos = np.zeros_like(x_pos)

# Bar heights (data values)
dx = np.ones_like(z_pos) * 1  # Bar width
dy = np.ones_like(z_pos) * 1  # Bar depth
dz = df.values.flatten()

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color='c', zsort='average')

ax.set_xlabel('Date')
ax.set_ylabel('Columns')
ax.set_zlabel('Values')
plt.title('3D Bar Chart for the DataFrame')
plt.show()
```

```python
# 安裝必要套件（如果尚未安裝）
# !pip install pandas statsmodels

import pandas as pd
import random
from statsmodels.formula.api import ols
import statsmodels.api as sm

# 原始數據
data = {
    'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81, 79, 84, 77, 80],
    'Method': ['A'] * 6 + ['B'] * 6 + ['C'] * 6
}

# 將數據轉換為 DataFrame
df = pd.DataFrame(data)

# 擴充數據集至 100 筆
new_data = []
methods = ['A', 'B', 'C']
for _ in range(82):  # 新增 82 筆數據
    score = random.randint(70, 100)  # 隨機生成 70 至 100 的分數
    method = random.choice(methods)  # 隨機選擇方法
    new_data.append({'Score': score, 'Method': method})

# 合併原始數據與新數據
df_expanded = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)

# 使用 statsmodels 進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df_expanded).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 輸出 ANOVA 表
print("擴充後數據集：")
print(df_expanded)
print("\nANOVA 表：")
print(anova_table)
```

```python
# 安裝必要套件（如需要）
# !pip install pandas matplotlib seaborn statsmodels

import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
import statsmodels.api as sm

# 原始數據
data = {
    'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81, 79, 84, 77, 80],
    'Method': ['A'] * 6 + ['B'] * 6 + ['C'] * 6
}

# 將數據轉換為 DataFrame
df = pd.DataFrame(data)

# 擴充數據集至 100 筆
new_data = []
methods = ['A', 'B', 'C']
for _ in range(82):  # 新增 82 筆數據
    score = random.randint(70, 100)  # 隨機生成 70 至 100 的分數
    method = random.choice(methods)  # 隨機選擇方法
    new_data.append({'Score': score, 'Method': method})

# 合併原始數據與新數據
df_expanded = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)

# 使用 statsmodels 進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df_expanded).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 輸出 ANOVA 表
print("ANOVA 表：")
print(anova_table)

# 設置樣式
sns.set(style="whitegrid")

# 1. 箱型圖 (Boxplot): 各方法的分數分布
plt.figure(figsize=(8, 6))
sns.boxplot(x='Method', y='Score', data=df_expanded, palette='Set2')
plt.title('Score Distribution by Method (Boxplot)', fontsize=14)
plt.xlabel('Method', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.show()

# 2. 散佈圖 (Scatter Plot): 每筆數據的分數
plt.figure(figsize=(10, 6))
sns.stripplot(x='Method', y='Score', data=df_expanded, jitter=True, palette='Set3', alpha=0.7)
plt.title('Individual Scores by Method (Scatter Plot)', fontsize=14)
plt.xlabel('Method', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.show()

# 3. 長條圖 (Bar Plot): 每個方法的平均分數
plt.figure(figsize=(8, 6))
mean_scores = df_expanded.groupby('Method')['Score'].mean().reset_index()
sns.barplot(x='Method', y='Score', data=mean_scores, palette='Set1')
plt.title('Average Scores by Method (Bar Plot)', fontsize=14)
plt.xlabel('Method', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.show()
```
