```python
import requests
import json
import pandas as pd

# 修正 URL
url = 'https://data.kcg.gov.tw/dataset/6f29f6f4-2549-4473-aa90-bf60d10895dc/resource/30dfc2cf-17b5-4a40-8bb7-c511ea166bd3/download/lightrailtraffic.json'

# 获取数据
web = requests.get(url)

# 手动处理 UTF-8 BOM
data = json.loads(web.content.decode('utf-8-sig'))

# 将 JSON 数据转换为 DataFrame
df = pd.DataFrame(data)

# 查看 DataFrame 的前几行
print(df.head())

# 导出为 CSV 文件到 Colab
df.to_csv('/content/lightrailtraffic.csv', index=False)  # 保存到 Colab 的根目录

# 确保文件可以下载
from google.colab import files
files.download('/content/lightrailtraffic.csv')
```

```python
import requests
from bs4 import BeautifulSoup

# 获取网页内容
url = 'https://tw.yahoo.com/'  # 替换为目标网站
response = requests.get(url)

# 检查请求状态码
if response.status_code == 200:
    html = response.text
else:
    print(f"Failed to fetch page. Status code: {response.status_code}")
    html = ""

# 创建 BeautifulSoup 对象
soup = BeautifulSoup(html, "html.parser")

# 移除所有 <script> 和 <style> 元素
for script in soup(["script", "style"]):
    script.extract()

# 获取网页的纯文本内容
text = soup.get_text()

# 处理文本，去除多余空行和空格
lines = (line.strip() for line in text.splitlines())  # 去掉每行前后的空格
text_cleaned = '\n'.join(line for line in lines if line)  # 保留非空行

# 打印最终结果
print(text_cleaned)
```

```python
import requests
from bs4 import BeautifulSoup
import os
from google.colab import drive

# 挂载 Google Drive（可选，如果需要保存到云端硬盘）
drive.mount('/content/drive')

# 定义并创建保存图片的目录
save_dir = '/content/drive/MyDrive/Notebooks/download/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 请求网页内容
web = requests.get('https://www.ptt.cc/bbs/Beauty/M.1710593264.A.578.html', cookies={'over18': '1'})

# 使用 BeautifulSoup 解析网页内容
soup = BeautifulSoup(web.text, "html.parser")

# 找出所有的 img 标签
imgs = soup.find_all('img')

# 图片编号从 0 开始
name = 0

# 遍历每个 img 标签
for i in imgs:
    # 使用 get() 方法安全获取 src 属性，避免没有 src 时抛出异常
    img_url = i.get('src')

    # 如果 img 标签没有 src 属性，跳过
    if img_url:
        # 检查是否是相对路径，如果是，则拼接为完整 URL
        if img_url.startswith('/'):
            img_url = 'https://www.ptt.cc' + img_url

        # 请求图片内容
        jpg = requests.get(img_url)

        # 保存图片到指定目录
        with open(f'{save_dir}test_{name}.jpg', 'wb') as f:
            f.write(jpg.content)  # 写入图片内容

        # 图片编号增加 1
        name += 1
```

```python
import requests
from bs4 import BeautifulSoup

# 请求网页内容
url = 'https://invoice.etax.nat.gov.tw/index.html'
web = requests.get(url)
web.encoding = 'utf-8'

# 使用 BeautifulSoup 解析 HTML 内容
soup = BeautifulSoup(web.text, "html.parser")

# 尝试选择包含中獎號碼的标签
td = soup.select('.container-fluid')

# 确保 td 不为空，避免 IndexError
if td:
    # 提取中獎號碼
    td = td[0].select('.etw-tbiggest')

    # 确保 td 中包含需要的元素
    if len(td) >= 5:
        ns = td[0].getText().strip()  # 特別獎
        n1 = td[1].getText().strip()  # 特獎
        # 取出最後八位數，避免多余换行符
        n2 = [item.getText().strip()[-8:] for item in td[2:5]]

        print(ns)
        print(n1)
        print(n2)
    else:
        print("没有找到足够的中奖号码元素")
else:
    print("没有找到包含中獎號碼的容器")
```

```python
import requests

# 牌告匯率 CSV 網址
url = 'https://www.stats.govt.nz/large-datasets/csv-files-for-download/'

# 爬取網址內容
rate = requests.get(url)
rate.encoding = 'utf-8'  # 調整回應訊息編碼為 utf-8，避免亂碼

# 以文字模式讀取內容
rt = rate.text
rts = rt.split('\n')  # 使用「換行」將內容拆分成串列

# 讀取串列的每個項目
for i in rts:
    try:
        # 用逗號拆分成子串列
        a = i.split(',')

        # 確保該行有足夠的列數，避免索引錯誤
        if len(a) > 12:
            print(a[0] + ': ' + a[12])  # 取出第一個 ( 0 ) 和第十三個項目 ( 12 )
    except Exception as e:
        # 捕捉異常並打印錯誤信息
        print(f"Error processing line: {i}, error: {e}")
        continue
```
